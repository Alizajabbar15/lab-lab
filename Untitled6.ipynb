```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt

#def cubicinterpolation1():
x_pts = np.linspace(0,2*np.pi,10)  #10 equidistance coords from 0 to 10
y_pts = np.sin(x_pts)
x_vals = np.linspace(0,2*np.pi,50)  # 50 desired points

f= interpolate.interp1d(x_pts,y_pts,'cubic')
y_vals = f(x_vals)   #cubic interpolation

plt.plot(x_pts,y_pts,'o')   #plot known data points
plt.plot(x_vals,y_vals,'-x')  #plot interpolated points
plt.show()

```

```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt

#def cubicinterpolation():
x_pts = np.linspace(0,2*np.pi,10)#10 equidistant coords
y_pts = np.sin(x_pts)
x_vals = np.linspace(0,2*np.pi,50) # 50 desired points
splines = interpolate.splrep(x_pts,y_pts)
y_vals = interpolate.splev(x_vals,splines) #cubic iterpolation

plt.plot(x_pts,y_pts,'o') # plot known data points
plt.plot(x_vals,y_vals,'-x') #plot interpolation points
plt.show
```
